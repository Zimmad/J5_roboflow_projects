```python
import os
from pathlib import Path

def create_yolo_structure():
    # Define the root directory
    root = Path("./")

    # Define all directories to be created
    directories = [
        "datasets/SVS-1/train",
        "datasets/SVS-1/val",
        "datasets/SVS-1/test",
        "datasets/Syrinx-2/train",
        "datasets/Syrinx-2/val",
        "datasets/Syrinx-2/test",
        "configs/datasets",
        "configs/augmentations",
        "src/data",
        "src/trainers",
        "src/evaluation",
        "src/utils",
        "runs/svc_exp/yolo9_baseline",
        "runs/svc_exp/yolo11_medical",
        "runs/svc_exp/yolo26_heavy",
        "runs/syrinx_exp/yolo9_baseline",
        "results/plots",
    ]

    # Define all files to be created
    files = [
        "configs/datasets/svs.yaml",
        "configs/datasets/syrinx.yaml",
        "configs/augmentations/aug_baseline.yaml",
        "configs/augmentations/aug_default.yaml",
        "configs/augmentations/medical_light.yaml",
        "configs/augmentations/medical_heavy.yaml",
        "src/data/prepare_splits.py",
        "src/data/verify_dataset.py",
        "src/trainers/train_yolo.py",
        "src/trainers/sweep_runner.py",
        "src/evaluation/evaluate.py",
        "src/evaluation/metrics.py",
        "src/evaluation/visualize.py",
        "src/utils/config.py",
        "src/utils/results_aggregator.py",
        "results/summary.csv",
        "run_experiments.py",
        "train_single.py",
        "requirements.txt",
        "README.md",
    ]

    print(f"Creating project structure in: {root.resolve()}")

    # Create directories
    for folder in directories:
        path = root / folder
        path.mkdir(parents=True, exist_ok=True)
    
    # Create empty files
    for file in files:
        file_path = root / file
        # Ensure the parent directory exists (just in case)
        file_path.parent.mkdir(parents=True, exist_ok=True)
        # Create file if it doesn't exist
        if not file_path.exists():
            file_path.touch()
            print(f"Created file: {file}")
        else:
            print(f"File already exists: {file}")

    print("\n✅ Project structure generated successfully!")

if __name__ == "__main__":
    create_yolo_structure()
```
